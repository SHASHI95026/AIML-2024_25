```python
import pandas as pd

people = ['kiran', 'arun', 'vijay', 'varun','rahul','prem']
age = [25, 30, 35, 40,20,19]
height = [145, 151, 165, 173,160,170]
weight = [45, 55, 65, 75,60,75]

data = {'People': people, 'Age': age, 'Height': height, 'Weight': weight}
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Age'], df['Height'])
plt.title('Age vs Height')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(df['People'], df['Height'])
plt.title('Height of People')
plt.xlabel('People')
plt.ylabel('Height')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.hist(df['Weight'],color='black',edgecolor='k', bins=10, alpha=0.5)
plt.title('Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.show()
```
